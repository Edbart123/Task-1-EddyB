The data will be analyzed to identify trends, sales patterns, and customer behaviors.

```python
df = df.drop_duplicates()

print(df.shape)
```

# Task 3: Exploratory Data Analysis (EDA)

In this section, we analyze the dataset to discover patterns, trends, and business insights.

```python
print(df["Product"].value_counts())
```

```python
print(df["TotalPrice"].mean())
```

```python
print(df["PaymentMethod"].value_counts())
```

```python
import matplotlib.pyplot as plt

df["Product"].value_counts().head(10).plot(kind="bar")

plt.title("Top Selling Products")
plt.xlabel("Products")
plt.ylabel("Count")

plt.show()
```

```python
print(df["PaymentMethod"].value_counts())
```

```python
df["PaymentMethod"].value_counts().plot(kind="pie", autopct='%1.1f%%')

plt.title("Payment Method Distribution")

plt.ylabel("")

plt.show()
```

```python
print("Average Order Price:", df["TotalPrice"].mean())
```

```python
print("Highest Order Price:", df["TotalPrice"].max())
```

```python
print("Lowest Order Price:", df["TotalPrice"].min())
```

```python
print(df["OrderStatus"].value_counts())
```

```python
df["OrderStatus"].value_counts().plot(kind="bar")

plt.title("Order Status Distribution")

plt.xlabel("Status")
plt.ylabel("Count")

plt.show()
```

# EDA Insights

Insights discovered from the dataset:

- Some products are significantly more popular than others.
- Digital payment methods are frequently used by customers.
- Most orders were completed successfully.
- A few products generated higher sales revenue compared to others.
- Customer purchasing behavior shows clear sales trends.
